# Data Science Assignment With Python

As part of a job application I was asked to complete a Data Science Assignment. I was given some data (a training set and a test set) and a few instructions:
I had to use the training set to produce predictions for the test set - the method I used was up to me but the goal was to minimise the Mean Squared Error of the predictions.

My first step was to take a look at the data. This can be done easily in Python.

Import the data:

```python
import pandas as pd
x_train = pd.read_csv('C:/Users/Snow/Desktop/assignment/X_train3.csv')
y_train = pd.read_csv('C:/Users/Snow/Desktop/assignment/y_train3.csv')
```

Basic Graph of the data:

```python
yplt = y_train.plot.line()
```

```python
xplt = x_train.plot.line()
```

Initally the data looks like it could be time series data. My next step was to check the autocorrelation function and the partial autocorrelation function to see what sort of data I was dealing with.

```python
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf
plot_acf(y_train)
plt.show()
```

```python
from statsmodels.graphics.tsaplots import plot_pacf
#plot_pacf(y_train)
#plt.show()
```

From these plots it was clear that this data was not time series data so applying methods like ARIMA and GARCH wouldn't be appropriate - my data was more likely cross-sectional.

My next step was to use an OLS regression - my goal was to minimise the mean squared error from my predictions which is exaclty what OLS is good for.

```python
import numpy
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

#split the data into a training set and a test set so I can calcualte the MSE
X_train, X_test, Y_train, Y_test = train_test_split(x_train, y_train)

ols = LinearRegression()
ols.fit(X_train,Y_train.values.ravel())
olsout = ols.predict(X_test)
print("OLS MSE = " + str(mean_squared_error(Y_test, olsout)))
```

I also returned the R-Squared value, which was not very encouraging:

```python
print("OLS R-Squared: " + str(ols.score(X_train, Y_train)))
```

I wanted more detail from my OLS so I swapped to using the stats models library OLS which has the ability to return an 'EViews-like' summary (unlike sklearn).

```python
import statsmodels.api as sm

X_trainc = sm.add_constant(X_train)
model = sm.OLS(Y_train, X_trainc).fit()
model.summary()
```

```python
import statsmodels.graphics as smgraphics

figure = smgraphics.regressionplots.plot_fit(model, 1)
```

The OLS regression suggested that some of the features were not significant, which in the spirit of parsimony could be removed, however as the significance values were based on a linear model and the relationship here appeared to be highly non-linear I would not remove any, plus it would incur a loss of data.

Next I wanted proof that the linear form was not adequate. This means using the Ramsey RESET test.

Performing the Ramsey RESET test on the OLS output also suggested the linear form was not adequate:

```python
from statsmodels.stats import outliers_influence

print(outliers_influence.reset_ramsey(model))
```

The null here is that the linear form is adquate which can be rejected.

Now here is when things get a bit less rigourous. As the purpose of the exercise was to show that I could perform basic Data Science techniques and the person who had developed the test referred to the regressors as 'features', I guessed they wanted me to demonstrate that I could run some simple Machine Learning methods and compare them.

I went on to try some different techniques including a random forest and a neural network which didn’t seem to offer any improvement over OLS which I somewhat expected.

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.neural_network import MLPRegressor

#GPR
grp = GaussianProcessRegressor()
grp.fit(X_train,Y_train.values.ravel())
pred  = grp.predict(X_test)
print("GPR MSE = " + str(mean_squared_error(Y_test, pred)))

#MLP Regression
mlp = MLPRegressor(hidden_layer_sizes=(50,50,50), max_iter=1000)
mlp.fit(X_train,Y_train.values.ravel())
predictions = mlp.predict(X_test)
print("MLP MSE = " + str(mean_squared_error(Y_test, predictions)))

#Random Forest
rf = RandomForestRegressor(n_estimators=100)
rf.fit(X_train,Y_train.values.ravel())
outp = rf.predict(X_test)
print("RF MSE = " + str(mean_squared_error(Y_test, outp)))
```

None of these techniques managed to beat the OLS for Mean Squared Errors which I expected, it was more an execise to show I could write the code.
